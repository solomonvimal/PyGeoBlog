```python
import pandas as pd
import geopandas as gpd
import numpy as np
import folium
from folium.plugins import MarkerCluster
from folium import IFrame
import shapely
from shapely.geometry import Point
import unicodedata
import pysal as ps
import numpy as np
import base64
import numpy as np
#lons = [77.57067837440151, 80.1999, 77.57101899999999]; lats = [13.04360946323446, 12.616, 13.01564]
import base64
from folium.plugins import MarkerCluster

m = folium.Map([np.mean(lats), np.mean(lons)], zoom_start = 2)
width, height, resolution = 30, 3, 10

folium.plugins.ScrollZoomToggler().add_to(m)

marker_cluster = MarkerCluster().add_to(m)


for i, lat in enumerate(lats[0:600]):
    lon = lons[i]
    text = good_urls[i]
    icon = folium.Icon(color="green", icon="ok")
    html =  "<html><header><title>"+str(text)+"</title></header><body>Lat "+str(text)+ "\n" + str(lat) + ", Lon " + str(lon) +"</body></html>" 
    #'<img src="data:image/png;base64,{}">'.format
    iframe = IFrame(html, width=(width*resolution), height=(height*resolution)+90)
    popup = folium.Popup(iframe, max_width=2650)
    icon = folium.Icon(color="green", icon="ok")
    #marker = folium.Marker(location=[lat, lon], popup=popup, icon=icon)
    #marker.add_to(m)
    
    folium.Marker([lat, lon],
        popup=popup,
        icon=folium.Icon(icon='ok-sign'),
        ).add_to(marker_cluster)
    
folium.plugins.Fullscreen(
    position='topright',
    title='Expand me',
    ).add_to(m)

hm_wide = folium.plugins.HeatMap( zip(lats, lons, lats), 
                   min_opacity=0.2,
                   max_val=55,
                   radius=17, blur=15, 
                   max_zoom=1, 
                 )
m.add_child(hm_wide)
m
```
